````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelBinarizer
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
````

````python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/letter-recognition/letter-recognition.data"
col_names = ["letter"] + [f"feat_{i}" for i in range(16)]
data = pd.read_csv(url, header=None, names=col_names)
````

````python
X = data.drop("letter", axis=1).values
y = data["letter"].values
````

````python
encoder = LabelBinarizer()
Y = encoder.fit_transform(y)
````

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
````

````python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

````python
model = Sequential([
    Dense(128, activation='relu', input_shape=(16,)),
    Dense(64, activation='relu'),
    Dense(26, activation='softmax')
])
````

````python
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
````

````python
history = model.fit(X_train, Y_train, epochs=25, batch_size=32, validation_split=0.1)
````

````python
test_loss, test_acc = model.evaluate(X_test, Y_test)
print(f"Test Accuracy: {test_acc:.4f}")
````

### Program 6: Standardize feature values

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**What it does**

- Computes mean and std from training set only.
- Transforms training and testing features to roughly zero mean and unit variance.

**Why important**

- Improves optimization speed and stability for neural networks.
- Prevents data leakage by not fitting scaler on test set.

**Expected output**

- No print.
- After scaling, each feature in `X_train` is approximately centered around 0.

## Viva Quick Questions You Can Answer

1. **Why one-hot encoding is used?**
   - Because output has 26 classes and softmax + categorical crossentropy expects one-hot targets.

2. **Why scale features before training?**
   - To make gradient descent stable and faster.

3. **Why softmax in final layer?**
   - It converts logits into class probabilities that sum to 1.

4. **Why use `random_state=42`?**
   - Reproducible train-test split.

5. **Difference between validation and test set?**
   - Validation is used during training for monitoring/tuning; test is used once for final unbiased performance estimate.

6. **Why `categorical_crossentropy`?**
   - Standard loss for multi-class one-hot encoded targets.

7. **What does `history` store?**
   - Per-epoch training and validation metrics.

8. **How is final class predicted?**
   - Take `argmax` of 26 softmax outputs.

9. **What indicates overfitting?**
   - Training accuracy increases but validation/test accuracy stagnates or decreases.

10. **What improvements are possible?**

- Early stopping, dropout, learning-rate scheduling, hyperparameter tuning, and confusion-matrix analysis.
